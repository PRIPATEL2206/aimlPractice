```python
import pandas as pd
```

```python
df=pd.read_csv("csv/heart.csv")
df.head()
```

```python
print(df.Sex.unique())
print(df.ChestPainType.unique())
print(df.RestingECG.unique())
print(df.ExerciseAngina.unique())
print(df.ST_Slope.unique())
```

```python
{i:j for i,j in enumerate(df.Sex.unique())}
```

```python
df.Sex=df.Sex.map({j:i for i,j in enumerate(iterable=df.Sex.unique())})
df.ChestPainType=df.ChestPainType.map({j:i for i,j in enumerate(iterable=df.ChestPainType.unique())})
df.RestingECG=df.RestingECG.map({j:i for i,j in enumerate(iterable=df.RestingECG.unique())})
df.ExerciseAngina=df.ExerciseAngina.map({j:i for i,j in enumerate(iterable=df.ExerciseAngina.unique())})
df.ST_Slope=df.ST_Slope.map({j:i for i,j in enumerate(iterable=df.ST_Slope.unique())})
df.head()
```
