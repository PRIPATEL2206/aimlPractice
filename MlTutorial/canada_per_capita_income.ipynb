```python
import pandas as pd
from  sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("csv/canada_per_capita_income.csv")
df.head()
```

```python
df.columns=["year","perCapIncom"]
```

```python
df[["perCapIncom"]].plot()
```

```python
model=linear_model.LinearRegression()
model.fit(df[["year"]],df["perCapIncom"])
```

```python
plt.scatter(df[["year"]],df.perCapIncom,c="red",marker="*")
plt.plot(df.year,model.predict(df[["year"]]))
```

```python
model.predict([[2020]])
```
