```python
norm_df=pd.get_dummies(df,dtype="int")
norm_df
```

```python
from sklearn.linear_model import LinearRegression


model=LinearRegression()
model.fit(norm_df[["Mileage","Age(yrs)","Car Model_Audi A5","Car Model_BMW X5","Car Model_Mercedez Benz C class"]],norm_df["Sell Price($)"])
```

```python
model.predict([[45000,4,0,0,1],[86000,7,0,1,0]])
```

```python
model.score(norm_df.drop(columns=["Sell Price($)"]),norm_df["Sell Price($)"])
```
