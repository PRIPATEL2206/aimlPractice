```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
def count(df):
    return df.price.describe()
def plot(df):
    print(count(df))
    plt.hist(df.price)
```

```python
plot(df)
```

```python
selectedQ=df.price.quantile(.995)
selectedQ
```

```python
df=df[df.price<=selectedQ]
plot(df)
```

```python
df.price.quantile(.99)
```

```python
plot(df[df.price>650])
```

```python
# removing data from 650
df=df[df.price<650]
plot(df)
```
